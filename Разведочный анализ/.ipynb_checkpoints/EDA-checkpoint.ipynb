```python
import pandas as pd
import numpy as np
```

```python
#Загрузка датасета;
df = pd.read_csv('vgsales_1.csv')
df
```

```python
# Просмотр и исследование датасета;
df.info()
```

```python
# - Обработка датасета:
#  - Удаление хотя бы одного признака;
df = df.where(df['NA_Sales'] < 0.3).dropna()
df
```

```python
# - Замена значений;
df['Platform'] = df['Platform'].mask(df['Platform'] == 'SNES', 'Super Nintendo Entertainment System')
df['Genre'] = df['Genre'].mask(df['Genre'] == 'Sports', 'Sport Simulators')
```

```python
# Использование или описание простейшего подстановщика;
from sklearn.impute import SimpleImputer

# создание объекта-подстановщика
imputer = SimpleImputer(strategy='mean')

# заполнение пропущенных значений в столбце 'EU_Sales' средним значением
df['EU_Sales'] = imputer.fit_transform(df[['EU_Sales']])
df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

features = ['Rank']
# Удалить строки, содержащие NaN
df.dropna(inplace=True)

# Заменить NaN на среднее значение по столбцу
df.fillna(df.mean(numeric_only=True), inplace=True)

# Матрица признаков и целевых переменных
X = df[features]
y = df['Global_Sales']

# Данные на обучающую и тесты
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Моделька линейной регрессии 
model = LinearRegression()

# Обучение этой модельки на данных 
model.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt

# диаграммма рассеивания точек
plt.scatter(X_test['Rank'], y_test, color='gray')
# Построить прямую линию на основе предсказаний модели
y_pred = model.predict(X_test)
plt.plot(X_test['Rank'], y_pred, color='red', linewidth = 2)

plt.xlabel('Rank - рейтинг')
plt.ylabel('Global Sales - глобальные продажи')
plt.title('График линейной регрессии продаж игр')
plt.show()
```
